```python
inspection_df = pd.read_csv("CSV_1_29_20.csv")
inspection_df.head()
```

```python
inspection_df =  inspection_df.loc[inspection_df["City"] == "Portland"]
inspection_df = inspection_df.drop_duplicates('Name')
inspection_df
```

```python
inspection_df['Rating'] = ''
inspection_df['Review Count'] = ''
inspection_df['Price'] = ''
inspection_df.head()
```

```python
url = 'https://api.yelp.com/v3/businesses/search'
headers = {'Authorization': 'Bearer %s' % api_key}
params = { 'limit' : 50, 'offset' : 0}





for index, row in inspection_df.iterrows():
        address = row['Address']
        city =  row['City']
        params['location'] = f'{address},{city}'
        req=requests.get(url, params=params, headers=headers).json()
        inspection_df.loc[index, "Review_Count"] = req["businesses"][0]["review_count"]
        inspection_df.loc[index, "Rating"] = req["businesses"][0]["rating"]
        inspection_df.loc[index, "Price"] = req["businesses"][0]["price"]
        params['offset'] += 50
inspection_df.head() 
    
  


```

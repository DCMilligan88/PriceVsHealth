```python
# Dependencies
import json
import pandas as pd
import requests
from matplotlib import pyplot as plt 
from config import api_key
from pprint import pprint
```

```python
inspection_df = pd.read_csv('InspectionOnlyPortland.csv')
inspection_df = inspection_df.sort_values('Name')# = inspection_df.drop_duplicates('Name')
```

```python
inspection_df
```

```python
inspection_df['Rating'] = ''
inspection_df['Review Count'] = ''
inspection_df['Price'] = ''
inspection_df['Yelp Name'] = ''
inspection_df = inspection_df.reset_index()
del inspection_df['index']
```

```python
inspection_df
```

```python
url = 'https://api.yelp.com/v3/businesses/search'
headers = {'Authorization': 'Bearer %s' % api_key}
params = {'location': 'Portland'}
counter = 0
# iterating through inspection_df
for index, row in inspection_df.iterrows():
    
    #creating variables to call in our api search
    zipc = row['Zip']
    city_rest = row['City']
    address_rest = row['Address']
    term = row['Name']
    #params['location'] = f'{address_rest},{city_rest},{zipc}'
    params['term'] = term
    response = requests.get(url, headers=headers, params=params).json()
    
    #try to populate the rest of dataframe
    try:
        results = response['businesses']
        inspection_df.loc[index, 'Rating'] = results[0]['rating']
        inspection_df.loc[index, 'Review Count'] = results[0]['review_count']
        inspection_df.loc[index, 'price'] = results[0]['price']
        inspection_df.loc[index, 'Yelp Name'] = results[0]['name']
    except:
        print("error... skipping")
        
    
        
```
